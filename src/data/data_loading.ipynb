# Data loading and preprocessing

Below we will 
1.  within-dataset division - Load the raw metahate dataset and divide into a 80% train set and 20% test set. 
2. LODO - leave one dataset out. We will try to  remove one dataset- OLID out from the metahate text to domain.
3. Load teh VAD values for the comments

1. Within Dataset division

```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("../../data/raw/available_metahate 15.tsv", sep="\t")
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
train_df.to_csv("train.tsv", sep="\t", index=False)
test_df.to_csv("test.tsv", sep="\t", index=False)
```

2. LODO- leave one dataset out
we will leave OLID dataset out in the training data and leave only this dataset as the test data.

```python
import pandas as pd
from sklearn.model_selection import train_test_split

metahateDataset = pd.read_csv("../../data/raw/available_metahate 15.tsv", sep="\t")
olidDataset = pd.read_csv("../../data/raw/olid-training-v1.0.tsv", sep="\t")

tweet_col = "tweet"

filtered_metahate = metahateDataset[
    ~metahateDataset["text"].isin(olidDataset[tweet_col])
]


filtered_metahate.to_csv(
    "../../data/processed/lodo/train_no_olid.tsv",
    sep="\t",
    index=False
)
```
